# Simple X-Y line plot 

```python
############################################
### Generate and plot a simple sine wave ###
############################################
x = np.linspace(0, 10, 101)
sin_x = np.sin(x)

plt.plot(x,sin_x)  

plt.title('sin(x)')
plt.show()
```

# Adjusting Line colors

Here, we create different plots of time shifted sine wave.
Matplotlib Pyplot (by default) sets different colors in multiple plots

```python
##############################################################
### Generate multiple sin plots with different time shifts ###
##############################################################
x = np.linspace(0, 10, 101)

plt.plot(x, np.sin(x - 0))
plt.plot(x, np.sin(x - 1))
plt.plot(x, np.sin(x - 2))
plt.plot(x, np.sin(x - 4))
plt.plot(x, np.sin(x - 5))

plt.legend(['sin(x-0)', 'sin(x-1)', 'sin(x-2)', 'sin(x-3)', 'sin(x-4)', 'sin(x-5)'])
plt.show()
```

However, we can manually assign color to each plot using `color` argument inside the `plt.plot()` function.

```python
x = np.linspace(0, 10, 101)

plt.plot(x, np.sin(x - 0), color = 'red')  # specify color by name
plt.plot(x, np.sin(x - 1), color = 'blue')
plt.plot(x, np.sin(x - 2), color = 'green')
plt.plot(x, np.sin(x - 4), color = 'black')
plt.plot(x, np.sin(x - 5), color = 'gray')

plt.legend(['sin(x-0)', 'sin(x-1)', 'sin(x-2)', 'sin(x-3)', 'sin(x-4)', 'sin(x-5)'])
plt.show()
```

# Histograms

```python
data = np.random.randn(1000)

n,bins,_ = plt.hist(data);
plt.title('Histogram of data')
plt.show()

print ('Total no of bins: {}\n'.format(len(n)))
for i in range(n.shape[0]):
    print ('Bins from {:+2.2f} to {:+2.2f} = {:4d} counts'.format(bins[i], bins[i+1], int(n[i])))
```

```python
plt.hist(data, bins=30, normed=True, alpha=0.5, histtype='bar', rwidth=.8, color='steelblue', edgecolor='blue');
plt.show()
```

```python
plt.hist(data, bins=30, normed=True, alpha=0.5, histtype='stepfilled', rwidth=None, color='steelblue', edgecolor='blue');
plt.show()
```

# Denisty & Contour plots

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + x * y) * np.cos(x)
    #return  np.sin(x**2 + y**2) / (x**2 + y**2)
```

```python
x = np.linspace(0, 5, 101)
y = np.linspace(0, 5, 101)

X, Y = np.meshgrid(x, y) # Returns all possible combinations of X and Y
Z = f(X, Y)
```

```python
### Draw contour lines ###

# When single color is used, negative values are represented by dashed lines and positive values by solid lines
plt.contour(X, Y, Z, colors='black'); #
plt.colorbar();
plt.show()
```

```python
### Draw filled contours ###

plt.contourf(X, Y, Z, 20, cmap='RdGy'); # 'RdGy' short for Red-Gray
plt.colorbar();
plt.show()
```

```python
# Continuous color plot ###

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

```python
### Combination of imshow and contour ###

contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8) # labels a contour plaot
plt.imshow(Z, extent=[0, 5, 0, 5,], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar();
```

# Create multiple plots using subplots

<p>The <code>subplot()</code> command specifies <b>numrows, numcols, fignum</b> where <b>fignum</b> ranges <i>from 1 to numrows*numcols</i>. 
<br><br>
<code>subplot(211)</code> is identical to <code>subplot(2,1,1)</code>
</p>

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211) # 211 - 2 rows, 1 col, 1st subplot
plt.plot(t1, np.sin(2*np.pi*t1), 'bo', t2, np.sin(2*np.pi*t2), 'k')
plt.title('sin(x)')

plt.subplot(212) # 212 - 2 rows, 1 col, 2nd subplot
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.title('cos(x)')

plt.subplots_adjust(hspace=.4) # height reserved for space between plots

plt.show()
```

# Create multiple figures

<p>We can create multiple figure by using <code>figure()</code> calls with an increasing figure number. In addition, each figure can contain as many subplots as we wish.
<br><br>
By default, when we don't call <code>figure()</code> and <code>subplot()</code>, <font size="3" color="blue"><code><font size="4" color="blue">figure(1)</font></code> and <code><font size="4" color="blue">subplot(111)</font></code> will be created by default.</font></p>

```python
plt.figure(1, figsize=(8,5))# the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([-4, -5, -6])
plt.title('Fig 1 - Tough as 4,5,6')

plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default
plt.title('Fig 2 - Tough as 4,5,6')

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Fig 1 - Easy as 1, 2, 3') # subplot 211 title
plt.subplots_adjust(hspace=.45) # height reserved for space between plots

plt.show()
```

# 3D plots

```python
### a basic scatterplot in 3D ###
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import


import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
