# K-NN - Explained in detail

**Meet the data - `forge` dataset from `mglearn`**
- Small and synthetic (i.e. made-up) classification dataset
- Has 2 features 
- Has 2 classes (Class 0 & 1) 
- Has 26 data instances (samples) in total

```python
import mglearn
import matplotlib.pyplot as plt

X,y = mglearn.datasets.make_forge()
print ("X shape:{}".format(X.shape))

# Create a scatter plot to vizualise all data points in the datasset
mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.legend(["Class 0", "Class 1"])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

- One of the simplest & most used classification algo.
- It's a non-parametric, lazy learning algo.
    - non-parametric - No parameters are learned from the model
    - lazy algo - There is no explicit training phase involved

## One nearest neighbour

In its simplest version, the `k-NN` algorithm only considers exactly **one nearest neighbor**, *which is the closest training data point to the point we want to make a prediction for* & assigns its label to the test data. 

Figure below illustrates this for the case of classification on the `forge` dataset. Here, 3 new test data points are added (shown in `star`). For each of the test point, the closest training data point is linked. The prediction label of the test point will be that of the closest training point (*as shown by the star color*).

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

## Two nearest neighbours

Instead of considering only the closest neighbor, we can also consider an arbitrary
number, k, of neighbors. **When considering more than one neighbor, we use majority voting to assign a label**. 
- This means that for each test point, we count how many neighbors belong to class 0 and how many neighbors belong to class 1. 
- We then assign the class that is more frequent: in other words, the majority class among the k-nearest neighbors. 

The following figure uses 2 nearest neighours

```python
mglearn.plots.plot_knn_classification(n_neighbors=2)
```

## Three nearest neighbours

We can see that the prediction for the new data point at the top left is not the same as the prediction when we used only one or two neigbours.

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
plt.plot(8.2,3.5, 'ro')
```

While this illustration is for a `binary classification` problem, this method can be
applied to datasets with any number of classes `multiclass classification` and of higher dimension in contrast to two-dimensions in this case. For more classes, we count how many neighbors belong to each class and again predict the most common class (`majority voting`).

# K-NN - Training

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)

clf.fit(X_train, y_train)

print("Test set predictions: {}".format(clf.predict(X_test)))

print("Test set accuracy: {:.2f}".format(clf.score(X_test, y_test)))
```

## Analyzing K-NN Classifier

### Model complexity vs `n_neighbours`

With two-dimensional datasets, we can easily vizualise the prediction for all possible test points in the xy-plane. Let's color the plane according to the class that would be assigned to a point in this region. This lets us view the `decision boundary`.

<p style="color:green; font-size:110%;  border-left: 5px solid blue; padding-left: 10px;"><b>Decision boundary is the divide between where the algorithm assigns class 0 versus where it assigns class 1 to the new test data.</b></p>


Let's build KNN model for `K=1,3,9` on the `forge` dataset & plot the `decision boundary`.

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    # build the KNN model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    # create the scatter plot with decision boundary
    mglearn.plots.plot_2d_separator(knn, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(loc=3)
fig.suptitle("Decision boundaries created by the nearest neighbors model for different values of n_neighbor", y=1.1, fontsize=14)
plt.show()
```

From the figure on the left, we can see that **using a single neighbor results in a decision boundary that follows the training data closely**. **Considering more and more neighbors leads to a smoother decision boundary**. 

<p style="color:blue; font-size:110%;  border-left: 5px solid blue; padding-left: 10px;">A smoother boundary corresponds to a simpler model. In other words, using few neighbors corresponds to high model com‐ plexity (as shown on the left side of figure with K=1), and using many neighbors corresponds to low model complexity (as shown on the right side of figure with K=9).</p>


<p><b><u>Food for thought!</u></b></p>
*If we consider the extreme case where the number of neighbors is the number of all data points in the training set, which class the test set would belong to!*

### Model complexity vs Generalization

**Meet the data - `Breast Cancer` dataset from `sklearn`**

Let's see the connection between **model complexity** and **generalization** (*ability of the model to make accurate predictions on new, unseen dataset*).  We will do this on the real-world `Breast Cancer` dataset.

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
print (cancer['DESCR'][:1110] + '\n\t...')
```

 We begin by splitting the dataset into a training and a test set.

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, 
                                                    cancer.target, 
                                                    stratify=cancer.target, random_state=66)
print ("Training data shape:{}".format(X_train.shape))
print ("Test data shape    :{}".format(X_test.shape))
```

Then we evaluate training and test set performance with different numbers of neighbors.

*Note*: Concepts related to `Generalization, Overfitting & Underfitting` will be covered in a separate post.

```python
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 20
neighbors_settings = range(1, 21)

for n_neighbors in neighbors_settings:
    # intialize the knn model parameters
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    # build/ train the model
    knn.fit(X_train, y_train)
    # save training set accuracy
    training_accuracy.append(knn.score(X_train, y_train))
    # save test set (generalization) accuracy
    test_accuracy.append(knn.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, '--', label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.title("Comparison of training and test accuracy as a function of n_neighbors\n")
plt.legend()
plt.show()
```

While real-world plots are rarely very smooth, we can still recognize some of the characteristics of `overfitting` and `underfitting` (note that because considering fewer neighbors corresponds to a more complex model). 
- `overfitting` - Model works well on training set, not on test set as it's not able to generalize.
    - Model is too complex.
- `underfitting` - Model do badly on both training set as well as test set.
    - Model is too simple.

Considering a single nearest neighbor case in the above plot, the prediction accuracy on the training set is perfect (close to 1). But when more neighbors are considered, the model becomes simpler and the training accuracy drops.  The test set accuracy for using a single neighbor is lower than when using more neighbors, indicating that using the single nearest neighbor leads to a model that is too complex. On the other hand, when considering  10 neighbors, the model is too simple and performance is still worse. The worst performance is around 88% test accuracy (for `n_neighbour=2`), which might still be acceptable.

<p style="color:blue; font-size:110%;  border-left: 5px solid blue; padding-left: 10px;">The best performance is somewhere in the middle, using around six neighbors <code>n_neighbour=6</code> with a test accuracy of <code>0.937</code>.</p>

<p style="color:red;">Keep in mind that best <code>n_neighbours</code> vary from dataset to dataset & even within same dataset of different training/ test data size.</p> 

```python
print ('--- Based on test set/ generalization ---\n')

index = test_accuracy.index(max(test_accuracy))
print ('Best n_neighour  = {} | Training accuracy = {:0.3f} | Test accuracy = {:0.3f}'.
       format(neighbors_settings[index], training_accuracy[index], test_accuracy[index]))

index = test_accuracy.index(min(test_accuracy))
print ('Worst n_neighour = {} | Training accuracy = {:0.3f} | Test accuracy = {:0.3f}'.
       format(neighbors_settings[index], training_accuracy[index], test_accuracy[index]))
```

There is also a `regression` variant of the k-nearest neighbors algorithm.

**Meet the data - synthetic `wave` dataset from `mglearn`**
- Small and synthetic (i.e. made-up) regression dataset
- Has only 1 feature
- Continuous targer variable (or response or output)

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Input Feature")
plt.ylabel("Output Target")
plt.title("Plot of the wave dataset, with the x-axis showing the feature and the y-axis showing the regression target", y=1.1)
plt.show()
```

## One nearest neighbour

Again, let’s start by using the single nearest neighbor, this time using the `wave` dataset. We’ve added three test data points as `green stars` on the x-axis. The prediction using a single neighbor is just the target value of the nearest neighbor which are shown as `blue stars`.

```python
mglearn.plots.plot_knn_regression(n_neighbors=1)
plt.title('Predictions made by 1-NN regression on the wave dataset', y=1.1, fontsize=14)
plt.xlabel('Input Feature')
plt.show()
```

## Three nearest neighbours

Again, we can use more than the single closest neighbor for regression. When using multiple nearest neighbors, the prediction is the average, or mean, of the relevant neighbors. We can see that the prediction target for the new data point at the bottom left `green star` is not the same as the prediction when we used only one neigbour.

```python
mglearn.plots.plot_knn_regression(n_neighbors=3)
plt.title('Predictions made by 3-NN regression on the wave dataset', y=1.1, fontsize=14)
plt.xlabel('Input Feature')
plt.show()
```

# K-NN Training

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

# split the wave dataset into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print ("Training data shape:{}".format(X_train.shape))
print ("Test data shape    :{}".format(X_test.shape))

# Set the number of neighbors to 3 & fit the KNN model on training set
reg = KNeighborsRegressor(n_neighbors=3).fit(X_train, y_train)
```

```python
# Make predictions on test set
y_predict = reg.predict(X_test)
print("Test set predictions:\n{}".format(y_predict))
```

```python
print ('First row  : Target value: {:0.3f} | Predicted value: {:0.3f}'.format(y_test[0], y_predict[0]))
print ('Second row : Target value: {:0.3f} | Predicted value: {:0.3f}'.format(y_test[1], y_predict[1]))
print ('Third row  : Target value: {:0.3f} | Predicted value: {:0.3f}'.format(y_test[2], y_predict[2]))
```

<p>For <code>Classificaiton</code> problems, we used <code>accuracy</code> as our evaluation metric. However for regression, we can't use accuracy. There are still many metrics to measure the regression model performance & by default, `score` method in `sklearn` returns <code>R<sup>2</sup></code> score.</p>

<p>The <code>R<sup>2</sup></code> score, also known as the <b>coefficient of determination</b>, is a measure of goodness of a prediction for a regression model, and yields a score between 0 and 1. A value of 1 corresponds to a perfect prediction</p>

`R^2 = (1 - u/v)`
where
- `u` is the residual sum of squares `((y_true - y_pred) ** 2).sum()`
- `v` is the total sum of square `((y_true - y_true.mean())**2)`

```python
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

A R^2 score of 0.83 indicates a relatively good model fit.

## Analyzing K-NN Regressor

### Model complexity vs `n_neighbours`

For our one-dimensional `wave` dataset, we can see what the predictions look like for all possible input feature values. To do this, let's create a test dataset consisting of many points on the line:

```python
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle("Comparing predictions made by nearest neighbors regression for diﬀerent values of n_neighbors", y = 1.1, fontsize=14)
# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
```

```python
for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using n_neighbours as 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} | test score: {:.2f}".format(
        n_neighbors, reg.score(X_train, y_train), reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target",  "Test data/target"], loc="best")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

n_neighbors = 15

# import some data to play with
iris = datasets.load_iris()

# we only take the first two features. We could avoid this ugly
# slicing by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

for weights in ['uniform', 'distance']:
    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("3-Class classification (k = %i, weights = '%s')"
              % (n_neighbors, weights))
```

# Train & Test Data

<p style="color:red; font-size:120%;  border-left: 5px solid blue; padding-left: 10px;">Never train on test data</p>

In Machine Learning, the first thing is to split the dataset into 2 subsets:
- **training set** - Set used to train the model.
- **test set** - Set used to test the trained model.

**Training set**
- This dataset will be used to build our model (i.e. to learn or to fit the parameters) .
- *e.g. weights in a binary logistic classifier.*

**Test set**
- This dataset is used to asses the model's performance.
- It's independent of the training dataset, but that follows the same probability distribution as the training datase (i.e. don't pick a test set with different characteristics than the training set).

**Validation set**
- This dataset is used to tune the hyperparameters (i.e. the architecture) of our model to avoid overtfitting.
- *e.g. In artificial neural networks, a hyperparameter is, for example, the number of neurons/nodes in each layer.*

<p>For further reading, refer this <a href = "https://en.wikipedia.org/wiki/Training,_test,_and_validation_sets" target="_blank"> Training, Test and Validaiton Sets - Wiki Page</a></p>

*We will see more about this splitting the data sets in later posts*

In our first ML model, we will not be tuning any hyperparameters and so, there's no need for a validation set.

Using `scikit-learn` built-in function `train_test_split`, we can **shuffle** the dataset and split them into `train set (75%)` and `test set (25%)`.

We set the `random_state=0` inside the `train_test_split()` to make sure we get the same output if we run the function multiple times (for repeatability). 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(iris_dataset['data'], 
                                                    iris_dataset['target'], 
                                                    random_state=0)
```

The `train_test_split()` function extracts 75% of data together with corresponding labels as training set and the rest as test set. Using a test set containing 20-25% of overall data is a good rule of thumb.

<p style="color:gray; font-size:115%;  border-left: 5px solid blue; padding-left: 10px;">Shuffle the data before train-test splitting to make sure that the test set has similar characteristics as training set.</p>


Before splitting the data, `train_test_split()` function shuffles the dataset using a pseudorandom number generator. 
**Why should we shuffle?**
- If we don't shuffle, the test or train set might contain all data points from one target class (not a good generalization of entire data set). 
- A typical classifier will probably fail miserably (underfit/ overfit) & some classifiers might even throw errors during such case.

# Building our first model: K-NN

Let's build the actual machine learning model. There are many classification algorithms in `scikit-leanr` and other python packages. We will use `k-nearest neighbours` classifier, which is easy to use & understand.

## K-NN - Quick Intro

- One of the simplest & most used classification algo.
- It's a non-parametric, lazy learning algo.
    - non-parametric - No parameters are learned from the model
    - lazy algo - There is no explicit training phase involved
---

Think of the figure shown below as a two class problem (`red` and `blue` points).
Our goal is to predict which target class the new test data - '`green`' point belongs to!

**Training**
- Building this model just involves storing the training data set (`red` and `blue` points).

**Testing**
- To make a prediction for a new test data (`green` point), the algo finds the point in the training set closest to the new data point (*it will compare with each `blue` & `red` point*).
- Then, it assigns label of the training point to the new data point

<p><img src="https://prasanth-ntu.github.io/html/ML-Course/Intro-to-ML-with-Python/images/KNN-wikipedia.png" 
	alt="KNN-wikipedia" title="KNN-wikipedia" style="max-width:300px;"></p>

The `k` in `k-nearest neighbour` define the number of neighbours to consider before assigning a label to the test data point.

- If `k=1`, it will use only the closest neighbour to the new data point
    - '`green`' point will be classified as '`red`' label as '`red`' data point is the closest to it
- If `k=2`, it will use 2 closest neighbours to the new data point
    - '`green`' point will still be classified as '`red`' label as the 2 data points closest to it are '`red`'
- If `k=3`, it will use 3 closest neighbours to the new data point
    - '`green`' point will still be classified as '`red`' label as 2 among 3 data points closest to it are '`red`'
- If `k=5`, it will use 5 closest neighbours to the new data point
    - '`green`' point will now be classified as '`blue`' label as 3 among 5 data points closest to it are '`blue`'
    
*We will revisit KNN in a much greater depth in the future posts.*

---
`sklearn.neighbours` module contains the `KNeighborsClassifier` algorithm.
- To use the algo, we need to define the `n_neighbours` parameter. We will set it to `1` for now. 
- We will store the object returned in variable named `knn`.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

To build the model on training set, we call the `knn` object with `fit()` method which takes 2 arguments:
- Training data
- Training labels/ classes

```python
knn.fit(X_train, Y_train)
```

In addition to `n_neighbours` parameter in `KNeighborsClassifier`, there are many other parameters which is returned by `fit()` method as seen above. We dont have to worry about other parameters now.
*The important parameters will be covered in future posts.*

# Making predictions

Now that we have built our first ML model using KNN on Training data set, we can now make predictions on test set to see how well our model performs on new unseen dataset. 

```python
print ('First row :', 'Measurements', X_test[0], ' | Target Label: ', Y_test[0])
print ('Second row:', 'Measurements', X_test[1], ' | Target Label: ', Y_test[1])
```

The first row in `test` data set belong to `target class 2` which is `virginica`. The second row in `test` data set belong to `target class 1` which is `versicolor`.
Let's see whether our model is able to predict it properly.

To make prediction, we need to call `predict()` function of the `knn` object with data to be predicted as input argument.

```python
y_pred = knn.predict(X_test[0:2])
print ('Predicted label for test data - First row :', y_pred[0])
print ('Predicted label for test data - Second row:', y_pred[1])
```

This matches with our original target of the test set. Seems like our ML model have learnt quite well.


# Accuracy of our model
Well, we can't be sure of how well our model has learnt unless we test it on the entire `test data set`

```python
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
```

There are many metrics to measure the model performance & we will use `Accuracy` as our eval metric. 

<p style="color:gray; font-size:115%;  border-left: 5px solid blue; padding-left: 10px;">Accuracy is fraction of flowers for which the right species was predicted.

`accuracy = no of correct predictions / total no of dataset`

```python
import numpy as np
accuracy = np.mean(y_pred == Y_test)
print ("Test set accuracy: {:.4f}".format(accuracy))
```

To calculate accuracy, we need to call `score()` function with two input arguments.
- data to be predicted
- original target label 

```python
print ("Test set accuracy: {:.4f}".format(knn.score(X_test,Y_test)))
```

A test accuracy of 97% means we made right predictions for 97% of our flower species on test set. We can safely assume that our model to be correct 97% of time for new datasets. For a hobbist like me, this level of accuracy is trust worthy enough. 
Nevertheless, in coming posts, we will see how to improve the performance and limitations/ conditions are in tunining our model.
