## Supervised Machine Learning Algos

### Neural Networks (Deep Learning)

"`Deep Learning`" - Family of algorithms known as neural networks with more hidden layers. <br>
Here, we will discuss simple mehtods, namely
- **`Multilayer perceptrons (MLPs)` for classificaiton and regression**.
- MLPs are known as feed-forward neural networks or simply neural networks

<p><u><b>Linear Regressor</b></u></p>
- Prediction by a linear regressor (y is weighted sum of input features) is given by
    -<p><pre>y = w[0]x[0] + w[1]x[1] + w[2]x[2] + w[3]x[3] + ...  </pre></p>

In the figure below ...

- Each node on the left represents `input feature`
- Connecting lines represent the `learned coefficients/ weights`
- Node on the right represents the output which is a weighted sum of inputs.

```python
import mglearn
print ('---Visualization of logistic regression ---')
display(mglearn.plots.plot_logistic_regression_graph())
```

<p><u><b>TNe NN model</b></u></p>

- MLPS are generalizations of linear models in multiple stages making a decision.
- MLP is the process of computing weighted sums repeated multiple times as shown below
    - x - input features
    - h - hidden layer
    - y - output features
    - w - Weights between input x  & hidden layer h
    - v - Weights between hidden layer h and output y
    
<p><pre>
h[0] = non-linearity-fn(w[0,0]*x[0] + w[1,0]*x[1] + w[2,0]*x[2] + w[3,0]*x[3])
h[1] = non-linearity-fn(w[0,1]*x[0] + w[1,1]*x[1] + w[2,1]*x[2] + w[3,1]*x[3])
h[2] = non-linearity-fn(w[0,2]*x[0] + w[1,2]*x[1] + w[2,2]*x[2] + w[3,2]*x[3])

y[0] = v[0]*h[0] + v[1]*h[1] + v[2]*h[2]
</pre></p>

```python
import mglearn
print ('---Visualization of MLP with single hidden layer ---')
display(mglearn.plots.plot_single_hidden_layer_graph())
```

The above model has lot more coefficients (weights) to learn than a simple logisitc regression. <br>

Why there is `non-linearity-fn` after weighted sum from input to hidden layer?
- Computing a weigthed sum across multiple layers is same as computing one weighted sum.
- To make the NN model more powerful than a linear model, we apply a non-linear function to it
    - e.g. `relu - Rectified linear unit` - Cuts off value below zero
    - e.g. `tanh - Tangens hyperbolicus` - Saturates to 01 for low values and +1 for high values
    - e.g. `logisitc sigmoid function` - Range from -1 to 1

These non linear functions allows the NN to learn much more complicated function than a linear model could.

<a id = 'figure 2-46 non-linear activation functions'></a>

```python
import numpy as np
import matplotlib.pyplot as plt
import mglearn

line = np.linspace(-3,3,100)
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(line, np.tanh(line), label="tanh")
plt.plot(line, np.maximum(line, 0), label="relu")
plt.plot(line, 1/(1+np.exp(-line)), label="logisitc sigmoid")
plt.title('Non linear activation functions', fontsize=15)
plt.legend(loc='best'); plt.xlabel('x'); plt.ylabel('Non linear functions'); plt.grid(); plt.xlim([-2,2]); plt.ylim([-1,2])
plt.show()
```

#### First MLP NN model - Two Moons dataset
- Let's apply `MLPClassifier` to `two_moons` dataset
- By default, `MLPClassifier` uses 100 hidden nodes
- The solve used is `lbfgs` which will be discussed later

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons   # dataset consists of 2 halfmoon shapes, 
                                          # with each class consisting of 75 2D data points 
    
X,y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y, random_state =42)

# using 1 hidden layer, 100 neurons, lbfgs solver, relu activtion 
mlp = MLPClassifier(solver='lbfgs', random_state=0).fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.title('Decision boundary learned by NN with 100 hidden neurons on the given 2D dataset', fontsize=16)
plt.xlabel("Feature 0", fontsize=15); plt.ylabel("Feature 1", fontsize=15); plt.show()
```

#### Parameter Tuning
Tuning the parameters allow us to control the complexity of neural network thereby influencing the performance and accuracy.

The commonly tuned parameters are <i>
- No of neurons
- No of hidden layers
- Non-linearity activation functions
- l2 penalty for regularization
- solver for weight optimization
- learning rate
- batch size  
    </i>
    
   Let's tune some of these parameters and see how it affects the overall learning of the model.

##### No of neurons

From the above plot which uses 100 hidden neurons, it is clear than *our NN model learned a very non linear but ** smooth ** decision boundary*.<br>
Let's change the number of neurons in the hidden layer (*which will inturn reduce the complexity of the model*).

```python
# using 1 hidden layer, 10 neurons, lbfgs solver, relu activtion 
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10])
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.title('Decision boundary learned by NN with 10 hidden neurons on the given 2D dataset', fontsize=16)
plt.xlabel("Feature 0", fontsize=15); plt.ylabel("Feature 1", fontsize=15); plt.show()
```

##### No of hidden layers

From above plot, it's evident that with only 10 hidden neurons, the decision boundary is not smooth.
<p style="color:blue;">With a single hidden layer, this means the decision funciton will be made up of 10 straight line segments.</p><br>
The default nonlinearity function used is `relu`, as shown [here](#figure 2-46 non-linear activation functions). <br>
If we want a smoother decision boundary, we can add more neurons, add more hidden layers and try different non-linearity functions.

```python
# using 2 hidden layer, 10 neurons, lbfgs solver, relu activtion 
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10,10])
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.title('Decision boundary learned by NN with 2 hidden layers, 10 hidden neurons each on the given 2D dataset', fontsize=16)
plt.xlabel("Feature 0", fontsize=15); plt.ylabel("Feature 1", fontsize=15); plt.show()
```

##### Non-linearity activation functions

```python
# using 2 hidden layer, 10 neurons, lbfgs solver, tanh activtion 
mlp = MLPClassifier(solver='lbfgs', activation = 'tanh', random_state=0, hidden_layer_sizes=[10,10])
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.title('Decision boundary learned by NN with 2 hidden layers, 10 hidden neurons each on the given 2D dataset', fontsize=16)
plt.xlabel("Feature 0", fontsize=15); plt.ylabel("Feature 1", fontsize=15); plt.show()
```

##### l2 penalty for regularization
- Using `l2 penalty`, we can control the complexity of the network by shrinking the weights towars zero (similar to ridge regression in linear classifiers).
- The parameter for this in `MLPClassifier` is `alpha` and its default value is very low (for little regularization)

```python
# 2 hidden layers, 2 different hidden neurons [10, 100], 4 different reg term [0.0001, 0.01, 0.1, 1]
print ('Decision functions for different numbers of hidden units and alpha parameters')
fig, axes = plt.subplots(2,4, figsize=(20,8))
for axx, n_hidden_nodes in zip(axes, [10,100]):
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
        mlp = MLPClassifier(solver='lbfgs', random_state=0, 
                            hidden_layer_sizes=[n_hidden_nodes], 
                            alpha=alpha)
        mlp.fit(X_train, y_train)
        mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)
        mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train, ax=ax)
        ax.set_title("n_hidden=[{},{}]\nalpha={:.4f}".format(
                    n_hidden_nodes, n_hidden_nodes, alpha))
```

Another important property of NN is that their weights are randomly initialized before learning is started, which affects the model learned. That mneas even with same parameters, we can obtain different results using different randome seeds. This effect might affect the accuracy of smaller networks. 

A common way to adjust the parametes in neural network is to first create a network that is large enough to overfit to make sure that training data can be learned. Once learned, either shrink the network or increase the `alpha` parameter to add regularization to improve generalization.

- The 4 parameters we tuned till now define the model we want to learn
- *How* to learn the model or what algorithm to use to learn the parameters and adjust the weights ? 

##### Solver
- The three solvers avial in MLP are 
    - `adam` - default and works will in most cases but is sensitive to scaling of data.
    - `lbfgs` - quite robust but might take long time on larger models or datasets
    - `sgs` - more advanced option, which many deep learning researchers use. It comes with additional parameters that can be tuned for better resutls. 
    
For beginners, `adam` and `lbfgs` are recommended.

##### MLP on Breast Cancer dataset
To get a better understanding of NN on real-world data, lets apply `MLPClassifier` to Breast Cancer dataset.<br>
At first, lets start with default parameters


```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

print ('Shape of Train set:', X_train.shape)
print ('Shape of Test set :', X_test.shape)

mlp = MLPClassifier(random_state=42)
mlp.fit(X_train, y_train)

print ("Accuracy on train set: {:.2f}".format(mlp.score(X_train, y_train)))
print ("Accuracy on test set: {:.2f}".format(mlp.score(X_test, y_test)))
```

Neural networks also expect all input features to ideally have a mean of 0 and variance of 1. Let's rescale to fulfill these requirements. 

```python
# Rescalling inptu features to have mean of 0 and variance of 1
mean_on_train = X_train.mean(axis=0)
std_on_train = X_train.std(axis=0)

X_train_scaled = (X_train - mean_on_train)/std_on_train  # After scaling, mean = 0 and std = 1
X_test_scaled = (X_test - mean_on_train)/std_on_train    # Apply same transformation as on training
```

```python
mlp = MLPClassifier(random_state=42)
mlp.fit(X_train_scaled, y_train)

print ("Accuracy on train set: {:.4f}".format(mlp.score(X_train_scaled, y_train)))
print ("Accuracy on test set: {:.4f}".format(mlp.score(X_test_scaled, y_test)))
```

The results are better after scaling, but we got a warning that *the maximum number of iterations have is reached*. So, we should change the default iterations to higher number

```python
mlp = MLPClassifier(max_iter = 1000, random_state=42)
mlp.fit(X_train_scaled, y_train)

print ("Accuracy on train set: {:.4f}".format(mlp.score(X_train_scaled, y_train)))
print ("Accuracy on test set: {:.4f}".format(mlp.score(X_test_scaled, y_test)))
```

<p style="color:red;">Increasing the iterations only increased the training set performance, not the generalization on test set. </p>Let's decrease the complexity of the learned model by increasing the `alpha` parameter (from `0.000` to `1`) to add stronger **regularization of the weights**.

```python
mlp = MLPClassifier(max_iter = 1000, alpha = 1, random_state=42)
mlp.fit(X_train_scaled, y_train)

print ("Accuracy on train set: {:.4f}".format(mlp.score(X_train_scaled, y_train)))
print ("Accuracy on test set: {:.4f}".format(mlp.score(X_test_scaled, y_test)))
```

##### Interpreting what neural networks learnt
- Analysing what NN learned is trickeir than analyzing a tree based model or linear model. 
- One way to introspect what was learned is to look at the models weight.


The plot below shows the weights that were learnted connecting the input with 30 features to hidden layer with 100 neurons.
- The rows correspond to 30 input features
- The columns correspond to 100 hidden neurons

<p style="color:green;">Light colors represent large +ve values while dark colors represent -ve values.</p> 

```python
plt.figure(figsize=(20,10))
plt.imshow(mlp.coefs_[0], interpolation='none', cmap = 'viridis', aspect='auto')
plt.yticks(range(30), cancer.feature_names); 
plt.title('Heat map of 1st layer weights in NN learned on Breast Cancer dataset')
plt.xlabel("Cols in weight matrix"); plt.ylabel("Input feature"); plt.colorbar(); plt.show()
```

From the heatmap, we can infer that features that have very small weights for all of the hidden units are "less important" to the model.
- e.g. of "less important features"- Have relatively low weight values comapred to other features
    - `mean  smoothness`
    - `mean  compactness`
- e.g. of "more important features"- Have relatively high weight values compared to other features
    - `worst radius`
    - `worst texture`
    - `worst perimeter`
    
Though we could also visualize the weights connecting the hidden layer to output, those are difficult to interpret.

#### NN Strength, Weakness & Complexity
**NN Strength**
- NNs reemerged as state-of-the-art models in many ML applications
- Main advantages is that
    - They are able to capture info contained in large datasets
    - Build incredibly complex models
    
Given enough data and computational resources with carefu lparameter tuning, NN often beat other machine learning algos for classification and regression.

**NN Weakness**
- They take longer time to train (particualry long and compelx ones)
- They require careful preprocessing of data features
    - They work well with "homogenours" data, where all features have similar meaning

<p style="color:red;">Tuning NN parameters is an art unto itself.</p>

**Estimating NN complexity**
- Two most important parameters are
    - No of hidden layers
    - No of neurons in each hidden layer (*often similar to no of input features*)
    
- **Model complxeity** - *Can be represented using number of weights or coefficients learned*
    - e.g. 1 - Binary classification with 100 features and one hidden layer of 100 neurons
        - Weights between input and hidden layer = `100x100 = 10000`
        - Weights between hidden and output layer = `100x1 = 100`
        - Total weights = `10000 + 100 = 10100`
    - e.g. 2 - Binary classification with 100 features and two hidden layer of 100 neurons each
        - Weights between input and first hidden layer = `100x100 = 10000`
        - Weights between first and second hidden layer = `100x100 = 10000`
        - Weights between second hidden and output layer = `100x1 = 100`
        - Total weights = `10000 + 10000 + 100 = 20100`
    - e.g. 3 - Binary classification with 100 features and one hidden layer of 1000 neurons 
        - Weights between input and first hidden layer = `100x1000 = 100000`
        - Weights between hidden and output layer = `1000x1 = 1000`
        - Total weights = `100000 + 1000 = 101000`
